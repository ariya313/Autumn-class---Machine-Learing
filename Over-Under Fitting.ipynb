# Imports

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
%matplotlib inline
```

```python
model = LinearRegression(normalize=True)
```

# creating data

```python
x = np.random.rand(200)*20
y = 2*x**4 - 2 * x**3 - 5 * x**2 - 5 * x - 5
y = y + np.random.normal(scale=4000,size=y.shape)
y = y.reshape(-1, 1)
x = x.reshape(-1, 1)
#print(x)
grade=[1,2,4,8,16] 
trainerror=[]
testerror=[]
```

```python
n = int(x.shape[0] * 0.8)
xtrain = x[:n]
ytrain = y[:n]
xtest = x[n:]
ytest = y[n:]
#print(xtrain)
#print('this is Ytrain',ytrain)
#print(xtest)
```

# Part - 1 - k = 1

```python
model.fit(xtrain,ytrain)
xline = np.arange(0,20, 0.1)
xline = xline.reshape(-1, 1)

plt.figure()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("K = 1")
plt.plot(xline,model.predict(xline),color='#9a40f5',label="Model")
plt.plot(xtrain,ytrain,'bo',label="Train")
plt.plot(xtest,ytest,'o',color='#33caba',label="Test")
plt.legend()
#print((np.mean((ytrain - model.predict(xtrain)) ** 2) / 2))
trainerror.append(np.mean((ytrain - model.predict(xtrain)) ** 2) / 2)
testerror.append(np.mean((ytest - model.predict(xtest)) ** 2) / 2)
#print(trainerror)

#print(train_score.shape)
train_sizes , train_score, test_score = learning_curve(model,xtrain,ytrain, cv=5)
plt.figure()
plt.xlabel("Train Size")
plt.ylabel("Score")
plt.title("Learning Curve : K = 1")
plt.plot(train_sizes, train_score.mean(axis=1),'*-',label="Train")
plt.plot(train_sizes, test_score.mean(axis=1),'*-',label="Test")
plt.legend()
```

# Part - 2 - k = 2

```python
#print(xtrain)
x_train=xtrain
x_train=np.hstack((x_train,xtrain**2))
#print(x_train.shape)

#print(x_train)
model.fit(x_train,ytrain)

xline = np.arange(0,20, 0.1)
xline = xline.reshape(-1, 1)
x_line=xline
x_line=np.hstack((x_line,xline**2)) 
plt.figure()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("K = 2")
plt.plot(xline,model.predict(x_line),color='#9a40f5',label="Model")
plt.plot(xtrain,ytrain,'bo',label="Train")
plt.plot(xtest,ytest,'o',color='#33caba',label="Test")
plt.legend()

trainerror.append(np.mean((ytrain - model.predict(x_train)) ** 2) / 2)
x_test=xtest
x_test=np.hstack((x_test,xtest**2)) 
testerror.append(np.mean((ytest - model.predict(x_test)) ** 2) / 2)

train_sizes ,train_score, test_score = learning_curve(model,x_train,ytrain, cv=5)
plt.figure()
plt.xlabel("Train Size")
plt.ylabel("Score")
plt.title("Learning Curve : K = 2")
plt.plot(train_sizes, train_score.mean(axis=1),'o-',label="Train")
plt.plot(train_sizes, test_score.mean(axis=1),'o-',label="Test")
plt.legend()
```
